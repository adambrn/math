#### f(x)=-12x^(4)*sin(cos(x))-18x^(3)+5x^(2)+10x-30

 1.Определить корни

```python
import sympy as sp
```

```python
x = sp.Symbol('x')
```

```python
f = -12 * x**4 * sp.sin(sp.cos(x)) - 18 * x**3 + 5 * x**2 + 10 * x - 30
```

```python
sp.solveset(f, x)
```

2.Найти интервалы, на которых функция возрастает

```python
result = sp.diff(f) 
print(sp.solveset(result>0))
```

Бесконечное множество
